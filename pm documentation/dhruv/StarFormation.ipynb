```julia
using DifferentialEquations
using Plots

function star!(du, u, p, t)
    du[1] = 1-u[1]-u[2]-k1*(u[2]^2)*u[1]
    du[2] = k1*(u[2]^2)*u[1]+k2*(u[2]^n)*(u[1]+u[2]-1)
    nothing
end

k1=8.75
k2=2.5
n=1
u0 = [0.25; 0.4]
tspan = (0.0, 100.0)
prob = ODEProblem(star!, u0, tspan)
sol = solve(prob);

plot(sol, idxs=[(1,2)])

```
